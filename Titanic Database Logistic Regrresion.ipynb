```python
titanic.drop('Cabin',axis=1,inplace =True) 
titanic.head(10)
```

```python
titanic.dropna(inplace =True)
```

```python
sns.heatmap(titanic.isnull(), yticklabels =False ,cbar =False) 
```

```python
titanic.isnull().sum() 
```

```python
pd.get_dummies(titanic['Sex'],dtype =int)
```

```python
sex = pd.get_dummies(titanic['Sex'],dtype =int,drop_first = True)
sex.head(5)
```

```python
em = pd.get_dummies(titanic['Embarked'],dtype =int,drop_first=True)
em.head()
```

```python
pclass = pd.get_dummies(titanic['Pclass'],dtype =int,drop_first=True)
pclass.head()
```

```python
titanic.head()
```

```python
titanic = pd.concat([titanic,pclass,em,sex],axis=1) 
titanic.head()
```

```python
titanic.drop(['PassengerId','Pclass','Sex','Ticket','Embarked'],axis=1,inplace=True)
titanic.head()
```

```python
titanic.drop(['Name'],axis=1,inplace=True)
titanic.head()
```

```python
#TRAIN DATA
titanic.head()
```

```python
X=titanic.drop(['Survived'],axis =1)
y=titanic['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X.columns = X.columns.astype(str)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)  
# test_size 0.3 means data split into 70:30 ratio
```

```python
from sklearn.linear_model import LogisticRegression
titanic.head()
```

```python
logmodel = LogisticRegression(max_iter=1000)   # By default LogisticRegression has iterations = 100 we inc i to 1000
```

```python
logmodel.fit(X_train,y_train)
```

```python
predict = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test,predict)
```
